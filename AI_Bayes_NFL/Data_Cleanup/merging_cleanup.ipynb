# Notebook for Scratchwork and EDA

This notebook was used to merge datasets and calculate Average Points Allowed and Average Points Scores (per game).

```python
import pandas as pd
```

```python
df = pd.read_csv("../Data/nfl_first.csv")
```

```python
df['Team'] = df['Team'].str.replace(r"[*+]", "", regex=True)
```

```python
df99 = df[df["Year"] >= 1999]
```

```python
# Ignore warnings, calculates avg pts for and avg pts againts
df99["Avg_Points_Allowed"] = df99["PA"] / (df99["W"] + df99["L"] + df99["T"])
df99["Avg_Points_Scored"] = df99["PF"] / (df99["W"] + df99["L"] + df99["T"])
```

```python
df99.head()
```

```python
nfl_1999_stats = pd.read_csv("../Data/clean_nfl_1999_stats.csv")
```

```python
stats_playoffs = pd.merge(nfl_1999_stats, df99, on=["Year", "Team"])
```

```python
stats_playoffs["Avg_offense_success_rate"] = (stats_playoffs["offense_success_rate_pass"] + stats_playoffs["offense_success_rate_run"]) / 2
```

```python
stats_playoffs.to_csv("../Data/NFL_Merged_Data.csv")
```
